```python
# CELL 1 — IMPORTS AND DATA PULL
# Full VIX regime + Greeks + OI + IV surface analysis

import numpy as np
import pandas as pd
import yfinance as yf
import warnings
from scipy.stats import norm
from scipy.optimize import brentq
from math import log, sqrt, exp
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import time

warnings.filterwarnings('ignore')

print("Pulling data from Yahoo Finance...")
print("This covers 2010-2026 — no limits, full quality\n")

# ── PULL DAILY DATA ───────────────────────────────────────────
START = "2010-01-01"
END   = "2026-06-28"

# SPY, QQQ, VIX
spy = yf.download("SPY", start=START, end=END, auto_adjust=True, progress=False)
qqq = yf.download("QQQ", start=START, end=END, auto_adjust=True, progress=False)
vix = yf.download("^VIX", start=START, end=END, auto_adjust=True, progress=False)

# VIX term structure
vix9d  = yf.download("^VIX9D",  start="2014-01-01", end=END, auto_adjust=True, progress=False)
vix3m  = yf.download("^VIX3M",  start="2014-01-01", end=END, auto_adjust=True, progress=False)
vix6m  = yf.download("^VIX6M",  start="2014-01-01", end=END, auto_adjust=True, progress=False)

# SKEW index
skew = yf.download("^SKEW", start=START, end=END, auto_adjust=True, progress=False)

# Put/call ratio proxy via PCALL ETF and PPUT ETF
# Also pull VVIX (vol of vol)
vvix = yf.download("^VVIX", start=START, end=END, auto_adjust=True, progress=False)

print(f"SPY:   {len(spy)} days  ({spy.index[0].date()} to {spy.index[-1].date()})")
print(f"QQQ:   {len(qqq)} days")
print(f"VIX:   {len(vix)} days")
print(f"VIX9D: {len(vix9d)} days")
print(f"VIX3M: {len(vix3m)} days")
print(f"VIX6M: {len(vix6m)} days")
print(f"SKEW:  {len(skew)} days")
print(f"VVIX:  {len(vvix)} days")
print("\nAll data pulled successfully")
```
